```python
def heat_flow_approx(f, x_subintervals, t_subintervals, nu, a, b, T):
    solution = []
    h = (b-a)/float(x_subintervals)
    x_array = np.arange(a,b,h)
    x_array = np.append(x_array, [b])
    k = (T)/float(t_subintervals)
    U_i = np.zeros((len(x_array),))
    for i, x in enumerate(x_array):
        U_i[i] = f(x) 
    lam = nu*k/h**2.
    A = np.diag([1-2*(lam)]*(x_subintervals+1)) + np.diag([lam]*(x_subintervals), k=-1) + np.diag([lam]*(x_subintervals), k=1)
    for m in range(t_subintervals+1):
        U_i = A.dot(U_i)
        solution.append(U_i)
    return x_array, solution
```

```python
fig=plt.figure(figsize=(18, 16), dpi= 80, facecolor='w', edgecolor='k')
def f(x):
    return 2*max((.2 - abs(x - .5), 0))
x_subintervals = 6
t_subintervals = 10
nu = .05
a = 0 
b = 1
T = .4
heat_flow_approx_obj = heat_flow_approx(f, x_subintervals, t_subintervals, nu, a, b, T)
X = heat_flow_approx_obj[0]
U_0, U_10 = heat_flow_approx_obj[1][0], heat_flow_approx_obj[1][10]
plt.plot(X, U_0, '-o', label="t=0")
plt.plot(X, U_10, '-o', label="t=.4")
plt.legend()
plt.show()
```

```python
def animate_obj(X, T, solutions):
    x = X
    y = solutions
    fig = plt.figure(figsize=(9.75, 8), dpi= 80, facecolor='w', edgecolor='k')
    ax = fig.add_subplot(111, xlim=(-12,12), ylim=(0, 1))
    #plt.axes(xlim=(-12,12), ylim=(0,1))
    line, = plt.plot([], [], 'o-', lw=2) 
    time_text = ax.text(0.02, 0.95, '', transform=ax.transAxes)
    #line, = plt.plot([], [])
    def animate(i):
        interval = (1./float(len(y)))
        line.set_data(x, y[i])
        time_text.set_text('time = %.3f' % (i*float(T)/float(len(y))))
        return line, time_text, 
    anim = animation.FuncAnimation(fig, animate, frames=len(y), interval=50, blit=True)
    plt.close(anim._fig)
    return anim
```

```python
def f(x):
    return max((1 - abs(x**2), 0))
x_subintervals = 140
t_subintervals = 70
nu = 1
a = -12 
b = 12
T = 1
heat_flow_approx_obj_2 = heat_flow_approx(f, x_subintervals, t_subintervals, nu, a, b, T)
X = heat_flow_approx_obj_2[0]
U = heat_flow_approx_obj_2[1]
animate_obj(X, T, U)
```

# Breaking the CFL Condition
[Explanation here](https://en.wikipedia.org/wiki/Courant%E2%80%93Friedrichs%E2%80%93Lewy_condition)


```python
def f(x):
    return max((1 - abs(x**2), 0))
x_subintervals = 140
t_subintervals = 66
nu = 1
a = -12 
b = 12
T = 1
heat_flow_approx_obj_2_counter = heat_flow_approx(f, x_subintervals, t_subintervals, nu, a, b, T)
X = heat_flow_approx_obj_2_counter[0]
U = heat_flow_approx_obj_2_counter[1]
animate_obj(X, T, U)
```

```python
def Crank_Nicolson(f, x_subintervals, t_subintervals, nu, a, b, T):
    solution = []
    h = (b-a)/float(x_subintervals)
    x_array = np.arange(a,b,h)
    x_array = np.append(x_array, [b])
    k = (T)/float(t_subintervals)
    U_i = np.zeros((len(x_array),))
    for i, x in enumerate(x_array):
        U_i[i] = f(x) 
    lam = nu*k/(2.*h**2.)
    A = np.diag([1-2*(lam)]*(x_subintervals+1)) + np.diag([lam]*(x_subintervals), k=-1) + np.diag([lam]*(x_subintervals), k=1)
    B = np.diag([1+2*(lam)]*(x_subintervals+1)) + np.diag([-lam]*(x_subintervals), k=-1) + np.diag([-lam]*(x_subintervals), k=1)
    for m in range(t_subintervals+1):
        U_i = np.linalg.solve(B, (A.dot(U_i)))
        solution.append(U_i)
    return x_array, solution
```

```python
fig = plt.figure(figsize=(16, 8), dpi= 80, facecolor='w', edgecolor='k')
def f(x):
    return max((1 - x**2, 0))
nu = 1
a = -12 
b = 12
T = 1

total_approx = 6
actual = Crank_Nicolson(f, 10*2**total_approx, 10*2**total_approx, nu, a, b, T)[1][-1]#This is U(x, 1)
h_array = [(1/(10.*(x**2.))) for x in range(1,total_approx)]
error_array = []

for i in range(1, total_approx):
    crank_nicolson_obj = Crank_Nicolson(f, 10*2**i, 10*2**i, nu, a, b, T)
    error_array.append(np.max(np.abs(actual[::2**(total_approx-i)] - crank_nicolson_obj[1][-1])))

h_array = np.array(h_array)
error_array = np.array(error_array)

    
plt.loglog(h_array,error_array,".-k",label="$E(h)$")
plt.loglog(h_array,h_array**(2.),"-r",label="$h^{\, 2}$")
plt.legend(loc="best")
plt.show()
```
